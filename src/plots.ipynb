```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
subjects = ['0108', '0109', '0110', '0111', '0112', '0113', '0114', '0115']
data_folder = '../data/'
sensor_data = data_folder + 'sensor_data/'
sensor_class = data_folder + 'sensor_class/'

for subject in subjects:
    # Load the data
    with open(sensor_class + subject + '.pkl', 'rb') as f:
        loaded_data = pickle.load(f)

    # The loaded_data variable now contains the data you saved
    sensor_pos_neg_scores, pos_permutation_scores, pos_pvalues, pos_feat_importance = loaded_data

    # load times_0180.npy
    times = np.load(sensor_data + 'times_' + subject + '.npy')

    # Get the 5 and 95 percentiles of the permutation
    pos_permutations99 = np.quantile(pos_permutation_scores, 0.99, axis=1)
    pos_permutations01 = np.quantile(pos_permutation_scores, 0.01, axis=1)

    plt.figure()
    plt.plot(times, sensor_pos_neg_scores, color='k')
    plt.fill_between(times, pos_permutations01,
                        pos_permutations99, color="orange", alpha=0.35)
    plt.hlines(0.50, times[0], times[-1], linestyle='dashed', color='k')
    plt.ylabel('Proportion classified correctly')
    plt.xlabel('Time (s)')
    plt.title('Classification of positive vs. negative trials\n subject ' + subject + ' - sensor space')
    ###
    plt.ylim(0.4, 0.7)

    x_position = times[-1] + 0.1
    y_position = 0.50  
    plt.text(x_position, y_position, "99% permutation interval", color="orange", ha="right", va="center", rotation=90)

    plt.savefig('../plots/sensor/' + subject + '.png', dpi=300)
    plt.show()
```
